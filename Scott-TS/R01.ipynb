```r
plot(a10, ylab = "$ million", xlab = 'Year', main = "Antibiotic drug sales")
```

### Data shows strong trend and there seems to be seasality as well

```r
seasonplot(a10, ylab = "$ million", xlab = "Year", 
          main = "Seasonal plot : antidiabetic drug sales",
          year.labels = TRUE, year.labels.left = TRUE, col = 1:20, pch = 19)
```

### Seasonal plot. On y-axis in different years with increasing sales. On the x axis we can see a trend with initially decreasing sales and then increasing.

```r
monthplot(a10, ylab = "$ million", xlab = "month", xaxt = "n",
         main = "Seasonal deviation plot : antidiabetic drug sales")
axis(1, at = 1:12, labels = month.abb, cex = 0.8)
```

### Horizontals in the plot are average sales across that month across all of those years. There are kinks across these years giving us details of data.
