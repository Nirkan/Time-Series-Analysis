```r
beer2 <- window(ausbeer, start = 1992, end = 2006)

beer2
```

```r
# Correlation vs Autocorrelation

lag.plot(beer2, lags = 9, do.lines = FALSE)
```

### Above we can see correlation plots for different lags. Lag4 and Lag8 has positive correlation.

```r
# Correlation vs Autocorrelation

Acf(beer2)
```

### This essentially represent same imformation as the plot above. Lags 2, 6 show negative correlation and Lags 4, 8.
