```python
import pandas as pd

df = pd.read_csv('aapl_no_dates.csv')
df.head()
```

```python
rng = pd.date_range(start="6/1/2017", end = "6/30/2017", freq = "B")
rng
```

```python
df.set_index(rng, inplace = True)
df
```

```python
%matplotlib inline
df.Close.plot()
```

```python
df["2017-06-01":"2017-06-10"]
```

```python
df["2017-06-01":"2017-06-10"].Close.mean()
```

```python
df.asfreq('D', method = 'pad')             # Padding method will also fill stock prices on weekends as well
```

```python
df.asfreq('H', method = 'pad')              # One can also fill hourly stock prices as well
```
